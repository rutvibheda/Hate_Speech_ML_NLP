```python
def build_freqs(tweets, ys):
    yslist = np.squeeze(ys).tolist()
    freqs_0 = {}
    freqs_1 = {}
    freqs_2 = {}
    for y, tweet in zip(yslist, tweets):
        for word in ast.literal_eval(tweet):
            if y==0:
                pair = word
                if pair in freqs_0:
                    freqs_0[pair] += 1
                else:
                    freqs_0[pair] = 1
            elif y==2:
                pair = word
                if pair in freqs_2:
                    freqs_2[pair] += 1
                else:
                    freqs_2[pair] = 1
            else:
                pair = word
                if pair in freqs_1:
                    freqs_1[pair] += 1
                else:
                    freqs_1[pair] = 1

    return freqs_0,freqs_1,freqs_2
```

```python
freqs_0,freqs_1,freqs_2 = build_freqs(df['stemmed'].values, df['class'].values)
```
